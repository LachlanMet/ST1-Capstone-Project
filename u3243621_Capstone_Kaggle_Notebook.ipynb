```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage
import skimage.color as skic
import skimage.filters as skif
import skimage.data as skid
import skimage.util as sku
%matplotlib inline


d=3
img_path_1 = '/kaggle/input/militaryaircraftdetectiondataset/crop/B1/044425a07b59ca9b0f128f4f54eacc9a_0.jpg'
img_1 = cv2.imread(img_path_1)
img_1_blur3 = cv2.GaussianBlur(skic.rgb2gray(img_1), (2*d+1, 2*d+1), -1)[d:-d,d:-d]
img_path_2 = '/kaggle/input/militaryaircraftdetectiondataset/crop/F35/000aa01b25574f28b654718db0700f72_0.jpg'
img_2 = cv2.imread(img_path_2)
img_2_blur3 = cv2.GaussianBlur(skic.rgb2gray(img_2), (2*d+1, 2*d+1), -1)[d:-d,d:-d]
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols=2, nrows=2, figsize=(10, 10))


ax1.imshow(img_1)
ax1.set_title('Original Image\n B1')
ax1.set_axis_off()

ax2.imshow(img_1_blur3)
ax2.set_title('Blured Image\n B1')
ax2.set_axis_off()

ax3.imshow(img_2)
ax3.set_title('Original Image\n F35')
ax3.set_axis_off()

ax4.imshow(img_2_blur3)
ax4.set_title('Blured Image\n F35')
ax4.set_axis_off()
```
